```python
datos = data['zralocp']
(mu, sigma) = norm.fit(datos)

# the histogram of the data
plt.figure(1, figsize=(12, 9))
n, bins, patches = plt.hist(datos, 25, normed=1, facecolor='b', alpha=0.4)

# add a 'best fit' line
y = mlab.normpdf( bins, mu, sigma)
l = plt.plot(bins, y, 'r--', linewidth=2)

plt.axvline(mu, color='red', linestyle= 'dashed', linewidth=3)

#plot
plt.xlabel('Scores', fontsize=15)
plt.ylabel('Probability', fontsize=15)
plt.legend(loc='upper right')
plt.title(r'$\mathrm{Histogram\ of\ Test\ Scores:}\ \mu=%.3f,\ \sigma=%.3f$' %(mu, sigma), fontsize=20)
plt.grid(True)

plt.show()
```
